# tf.Example数据类型
`Example`数据类型是`{"string":tf.train.Feature}`映射

`Feature`可以接受三种类型
- `tf.train.BytesList`
- `tf.train.FloatList`
- `tf.train.Int64List`

```python
def _bytes_feature(value):
    '''value为string类型或byte类型。返回BytesList类型数据'''
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
    '''value为float32类型或float64类型。返回FloatList类型数据'''
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
    '''value为bool类型或enum类型或int类型或uint类型。返回Int64List类型数据'''
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

```python
b_str=_bytes_feature(b'b_str')
u_str=_bytes_feature(u'u_str'.encode('utf-8'))

text_float=_float_feature(np.exp(1))
                          
t_int64 = _int64_feature(True)
int_int64=_int64_feature(1)
```

## 创建一条`tf.Example`

序列化样本

```python
def serialize_example(f0,f1,f2,f3):
#     数据字典
    feature={
        'bool_feature':_int64_feature(f0),
        'integer_feature':_int64_feature(f1),
        'str_featurer':_bytes_feature(f2),
        'flot_feature':_float_feature(f3)
    }
#     创建特征列表
    features = tf.train.Features(feature=feature)
#     创建样本实例
    example_proto=tf.train.Example(features=features)
    print(example_proto)
#     序列化样本
    seri_example=example_proto.SerializeToString()
    return seri_example    
```

## 使用tf.data编写TFRecord文件

### 将__数据__导入__数据集__

创建特征数据

```python
# the number of observations in the dataset
n_observations = int(1e4)

# boolean feature, encoded as False or True
feature0 = np.random.choice([False, True], n_observations)

# integer feature, random from 0 .. 4
feature1 = np.random.randint(0, 5, n_observations)

# string feature
strings = np.array([b'cat', b'dog', b'chicken', b'horse', b'goat'])
feature2 = strings[feature1]

# float feature, from a standard normal distribution
feature3 = np.random.randn(n_observations)
```

应用于数组，返回标量数据集

```python
feature_set=tf.data.Dataset.from_tensor_slices(feature1)
feature_set
```

应用于元组，返回元组数据集

```python
features_dataset = tf.data.Dataset.from_tensor_slices((feature0, feature1, feature2, feature3))
features_dataset
```

```python
for f0,f1,f2,f3 in features_dataset.take(1):
    print(f0.numpy())
    print(f0)
    print(f1)
    print(f2)
    print(f3)
```

使用`tf.data.Dataset.map`方法将函数应用每个元素

```python
def serialize_example_pyfunc(f0,f1,f2,f3):
#    特征字典
    feature={
        'bool_feature':_int64_feature(f0.numpy()),
        'integer_feature':_int64_feature(f1.numpy()),
        'str_featurer':_bytes_feature(f2.numpy()),
        'flot_feature':_float_feature(f3.numpy())
    }
#     创建特征
    features = tf.train.Features(feature=feature)
#     创建样本实例
    example_proto=tf.train.Example(features=features)
#     序列化样本
    seri_example=example_proto.SerializeToString()
    return seri_example    
```

```python
def tf_serialize_example(f0,f1,f2,f3):
#     print(f0)
#     f0 = f0.numpy()
#     f1=f1.numpy()
#     f2=f2.numpy()
#     f3=f3.numpy()
    tf_str=tf.py_func(serialize_example_pyfunc,(f0,f1,f2,f3),tf.string)
    return tf.reshape(tf_str,())
```

```python
def image_example(image_str,label):
    image_decode=tf.image.decode_jpeg(image_str)
    image_shape=image_decode.shape
    print(image_shape)
    print(image_decode)
    feature={
        'height':_int64_feature(image_shape[0]),
        'width':_int64_feature(image_shape[1]),
        'depth':_int64_feature(image_shape[2]),
        'label':_int64_feature(label),
        'image_raw':_bytes_feature(image_str)
    }
    fture=tf.train.Features(feature=feature)
    img_examle=tf.train.Example(features=fture)
    return img_examle
    
```

## 阅读TFRecord文件

```python
raw_img_dataset=tf.data.TFRecordDataset(tfrecord_path)
```

```python
img_fture_descrip={
    'height':tf.FixedLenFeature([],tf.int64),
    'width':tf.FixedLenFeature([],tf.int64),
    'depth':tf.FixedLenFeature([],tf.int64),
    'label':tf.FixedLenFeature([],tf.int64),
    'image_raw':tf.FixedLenFeature([],tf.string)
}

def _parse_img_func(ex_proto):
    return tf.parse_single_example(ex_proto,img_fture_descrip)
```

```python
parse_img_dataset=raw_img_dataset.map(_parse_img_func)
```

```python
for img_fture in parse_img_dataset:
    img_raw_fture=img_fture['image_raw']
    print(type(img_raw_fture))
    img_raw=img_raw_fture.numpy()
    print( type(img_raw))
    display.display(display.Image(data=img_raw))
```
